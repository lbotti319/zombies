```python
import numpy as np
from zombies import corona_basic, plot_it
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
```

```python
y_0 = np.array([100,1,0,0,0,0,0])
# args: (b, m, alpha, z, k)
# test_args with random values
params = np.random.uniform(size=(18,))
solution = solve_ivp(
    corona_basic,
    (0, 10),
    y_0,
    first_step=0.05,
    max_step=0.05,
    args=params
)
```

```python
y = solution.y.T
```

```python
y[:,[0,1,6]]
```

```python
plt.plot(solution.t, solution.y.T[:,[1,2,6]])
plt.xlabel('t')
plt.legend(['Susceptibles', 'Exposed', 'Removed'], shadow=True, loc='upper right')
plt.title('Population Changes Over Time')
plt.show()
```
